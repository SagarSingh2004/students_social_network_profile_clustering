# <center>Students Social Network Profile Clustering</center>

---

## Problem Statement
This project aims to segment high school students into distinct groups based on their social networking profile data using unsupervised machine learning techniques.

The objective is to identify hidden patterns in student interests, demographics, and online behavior using clustering algorithms such as K-Means, Hierarchical Clustering, and DBSCAN.

These insights can help in understanding student behavior, interest-based grouping, and trend analysis in social networking platforms.

---

## Data Overview
This dataset consists of a random sample of approximately 15,000 high school students who maintained profiles on a popular social networking platform between 2006 and 2009.

The data was collected by crawling user profiles and processed using text mining techniques to extract dominant interest-related words. It captures the frequency of 37 commonly occurring terms found in student profiles.

Along with interest-based features, the dataset also includes key demographic attributes such as graduation year, gender, age, and number of friends. These features provide a rich foundation for clustering and behavioral analysis.

This dataset is well-suited for exploring:

- Student interest segmentation
- Demographic profiling
- Behavioral pattern discovery
- Trend analysis in online social activity
